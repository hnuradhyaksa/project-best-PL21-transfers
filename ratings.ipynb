```python
import pandas as pd
import numpy as np

stats = pd.read_csv(r'\project-best-PL21-22-transfers\data\player_stats_2021.csv')
```

```python
drop_col = ['played_per90','pos']
stats.drop(drop_col, inplace=True, axis=1)

stats = stats.fillna(0)

temp_drop = stats[['player','squad']]
stats = stats.drop(['player','squad'], axis=1)
stats = stats/stats.max()
stats[['player','squad']] = temp_drop

stats_t = stats.melt(id_vars=['player','squad'])

stats_t['byplayer_rank'] = (stats_t.groupby('player')['value']
                      .rank(method='dense', ascending=False)
                      .astype(int))

selected = pd.Series(stats_t['byplayer_rank'].unique()).nsmallest(5).values
stats_t=stats_t[stats_t['byplayer_rank'].isin(selected)]
```

```python
drop_col = ['byplayer_rank']
stats_t.drop(drop_col, inplace=True, axis=1)
total_stats = stats_t.groupby(['player', 'squad'])['value'].mean().to_frame().reset_index()
```

```python
tfee = pd.read_csv(r'\project-best-PL21-22-transfers\output\recent_pl_transfer.csv')
stats_tfee = total_stats.merge(tfee, left_on=['player'], right_on=['player_name'], how='inner')
stats_tfee = stats_tfee.drop(['player_name','squad'], axis=1)
stats_tfee
```

```python
progression = pd.read_csv(r'\project-best-PL21-22-transfers\data\progression_multiplier.csv')
total_rating = stats_tfee.merge(progression, left_on='club_name', right_on='Squad', how='left')
total_rating['value'] = total_rating['value']*0.6
total_rating['Normalized'] = total_rating['Normalized']*0.4
total_rating['total'] = total_rating['Normalized']+total_rating['value']
total_rating
```

```python
total_rating.to_csv("ratings_vs_tfee.csv", index=True)
```
